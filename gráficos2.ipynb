```python
#Gr√°fico de ondas estacionarias para diferentes periodos
import numpy as np
import matplotlib.pyplot as plt
from scipy import constants as cte

z = np.arange(0.1,10,0.01)
E_io = 1
#Para el aire
beta = np.pi/2
w = beta/(cte.mu_0*cte.epsilon_0) 
T = (2*np.pi)/w
def E(z,t):
  return 2*E_io*np.sin(beta*z)*np.sin(w*t)

#Graficamos
fig, ax = plt.subplots(figsize=(7, 5), constrained_layout=True)
plt.plot(z, E(z,0), color='blue',label='t = 0')
plt.plot(z, E(z,T/8),label='t = T/8')
plt.plot(z, E(z,T/4),label='t = T/4')
plt.plot(z, E(z,3*T/8),label='t = 3T/8')
plt.plot(z, E(z,T/2),label='t = T/2')
#plt.axvline(x=wp, ymin=0.0, ymax=5,label="$w_p$",c='black', linestyle='--')
plt.ylabel('Amplitud')
plt.xlabel('z')
plt.title('Ondas estacionarias')
plt.legend()
plt.show()
```

```python
#Grafico de los coeficientes de Fresnel para interfaz aire-BK7
import numpy as np
import matplotlib.pyplot as plt

n1=  1         #aire
n2=  1.52         #BK7
def rll(Oi):
        return (n2*np.cos(Ot) - n1*np.cos(Oi))/(n1*np.cos(Oi) + n2*np.cos(Ot))
def rT(Oi):
        return (n2*np.cos(Oi)-n1*np.cos(Ot))/(n2*np.cos(Oi) + n1*np.cos(Ot))
def tll(Oi):
        return (2*n2*np.cos(Oi))/(n1*np.cos(Oi) + n2*np.cos(Ot))
def tT(Oi):
        return (2*n2*np.cos(Oi))/(n2*np.cos(Oi) + n1*np.cos(Ot))
#Graficamos
def A(Oi):
  return Oi/Oi-1
Ot=0.6
Oi=np.linspace(0,2*np.pi,1000)

    
plt.plot(Oi,rll(Oi),'brown')
plt.plot(Oi,tll(Oi),'r')
plt.plot(Oi,rT(Oi),'g')
plt.plot(Oi,tT(Oi),'y')
plt.plot(Oi,A(Oi),'black')
plt.legend(['$\Gamma_{||}$','$T_{||}$','$\Gamma_\perp$','$T_\perp$'],loc='lower right')
plt.xlabel("$theta_i$")
plt.ylabel("Coeficientes")
plt.title("Coeficientes de fresnel")
plt.xlim(0,np.pi/2)
plt.ylim(-1,1)
plt.grid()
plt.show()
```

```python
#Grafico de los coeficientes de Fresnel para interfaz aire-algo
import numpy as np
import matplotlib.pyplot as plt

n=1/1.5
def rll(Oi):
        return ((n**2-np.sin(Oi)**2)**(0.5) - n**2*np.cos(Oi))/(n**2*np.cos(Oi) +(n**2-np.sin(Oi)**2)**0.5)
def rT(Oi):
        return (np.cos(Oi)-(n**2-np.sin(Oi)**2)**0.5)/(np.cos(Oi) + (n**2-np.sin(Oi)**2)**0.5)
def tll(Oi):
        return (2*n*np.cos(Oi))/(n**2*np.cos(Oi) + (n**2-np.sin(Oi)**2)**0.5)
def tT(Oi):
        return (2*np.cos(Oi))/(np.cos(Oi) + (n**2-np.sin(Oi)**2)**0.5)
#Graficamos
def A(Oi):
  return Oi/Oi-1

Oi=np.linspace(0,2*np.pi,1000)

    
plt.plot(Oi,rll(Oi),'brown')
plt.plot(Oi,tll(Oi),'r')
plt.plot(Oi,rT(Oi),'g')
plt.plot(Oi,tT(Oi),'y')
plt.plot(Oi,A(Oi),'black')
plt.legend(['$\Gamma_{TM}$','$T_{TM}$','$\Gamma_{TE}$','$T_{TE}$'],loc='lower right')
plt.xlabel('$theta_i$')
plt.ylabel("Coeficientes")
plt.title("Coeficientes de fresnel SU8-aire")
plt.xlim(0,np.pi/2)
plt.ylim(-1,1)
plt.grid()
plt.show()
```

```python
#Grafico de las reflectancias (TE y TM) para reflexi√≥n interna y externa Aire y bk7
import numpy as np
import matplotlib.pyplot as plt

n=1.5
def RM(Oi):
        return (((n**2-np.sin(Oi)**2)**(0.5)- n**2*np.cos(Oi))/(n**2*np.cos(Oi) +(n**2-np.sin(Oi)**2)**0.5))**2
def RE(Oi):
        return ((np.cos(Oi)-(n**2-np.sin(Oi)**2)**0.5)/(np.cos(Oi) + (n**2-np.sin(Oi)**2)**0.5))**2

#Graficamos
def A(Oi):
  return Oi/Oi-1

Oi=np.linspace(0,np.pi,1000)

    
plt.plot(Oi,RE(Oi),'red')
plt.plot(Oi,RM(Oi),'red',linestyle='--')

plt.axvline(x=np.arctan(n), ymin=0.0, ymax=1,c='green', linestyle='--')
plt.axvline(x=np.arctan(1/n), ymin=0.0, ymax=1,c='brown', linestyle='--')
plt.axvline(x=0.74, ymin=0.0, ymax=1,c='black', linestyle='--')
n=1/n
plt.plot(Oi,RE(Oi),'blue')
plt.plot(Oi,RM(Oi),'blue',linestyle='--')
plt.legend(['$R_{TE(Externo)}$','$R_{TM (Externo)}$',"$\Theta_p=56^\circ$","$\Theta_p=34^\circ$","$\Theta_c=42^\circ$",'$R_{TE (Interno)}$','$R_{TM (Interno)}$'],loc='upper right')
plt.xlabel('$theta_i$')
plt.ylabel("Reflectancias")
plt.title("Reflectancias para los casos Interno y Externo")

plt.xlim(0,np.pi/2)
plt.ylim(-0.1,1)
plt.grid()
plt.show()
```

```python
#Grafico del cambio de fase en las reflexiones de luz
import numpy as np
import matplotlib.pyplot as plt

n=1/1.5
def phiTE(Oi):
        return -2*np.arctan(((np.sin(Oi)**2-n**2)**0.5)/(np.cos(Oi)))
def phiTM(Oi):
        return np.pi-2*np.arctan(((np.sin(Oi)**2-n**2)**0.5)/(n**2*np.cos(Oi)))
#Graficamos
def A(Oi):
  return Oi/Oi-1
def B(Oi):
  return np.pi*Oi/Oi
Oi=np.linspace(0,2*np.pi,1000)
Ai=np.linspace(0,0.74,1000)
x=np.linspace(0.59,0.74,1000)    
plt.plot(Oi,phiTE(Oi),'r')
plt.plot(Oi,phiTM(Oi),'green')
plt.axvline(x=np.arctan(n), ymin=0.0, ymax=1,c='green', linestyle='--')
plt.axvline(x=0.74, ymin=0.0, ymax=1,c='black', linestyle='--')

plt.plot(x,B(x),'green')
plt.plot(Ai,A(Ai),'red')
plt.legend(['$\phi{TE}$','$\phi{TM}$',"$\Theta_p=34^\circ$","$\Theta_c=42^\circ$"],loc='lower right')
plt.xlabel('$\Theta_i$')
plt.ylabel("Cambio de fase")
plt.title("Cambio de fase para la reflexi√≥n")
plt.xlim(0,np.pi/2-0.1)
plt.ylim(-4,4)
plt.grid()
plt.show()
```
