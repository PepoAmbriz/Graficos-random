```
#José Francisco Ambriz Gutiérrez
import numpy as np
import matplotlib.pyplot as plt

t=np.linspace(0,40,10000)
A=1
fase=np.pi/2
b=0.11
w0=1
w1=np.sqrt(w0**2-b**2)

x=A*np.exp(-b*t)*np.cos(w1*t-fase)
v=-A*np.exp(-b*t)*(b*np.cos(w1*t-fase)+w1*np.sin(w1*t-fase))
u=w1*x
z=b*x+v

#X(t) del mov amortiguado
plt.plot(t,x)
plt.xlabel("t")
plt.ylabel("x(t)")

f = plt.figure(figsize=(10.9,5))
uw = f.add_subplot(121)
ax = f.add_subplot(122)
#z(u) lissajous graficos de osciladores
uw.plot(u,z,color="blue")
#v(x) lissajous
ax.plot(x, v,color="red")
ax.grid(1)
uw.grid(1)
```

```
#Gráficos de distribución binomial
import numpy as np
import matplotlib.pyplot as plt
from math import factorial

N=12
for N1 in range(0,N+1):
  wn=(factorial(N)*(2/3)**(N1)*(1/3)**(N-N1))/(factorial(N1)*factorial(N-N1))
  plt.plot(N1/N,wn,'.')

plt.xlabel("N1/N")
plt.ylabel("Pn")
plt.show()
```

```
#Graficos de \alpha y \beta del material Si y SiO_2 en un dispositivo fotónico.
import numpy as np
import matplotlib.pyplot as plt

#Si
#sig= 1.67*10**(-2)
#eps=11.7*8.85*10**(-12)
#mu=1/((74812)**2*eps)
#SiO2
sig= 10**(-18)
eps=3.45*10**(-11)
mu=1.26*10**(-7)

def alpha(w):
        return w*(mu*eps/2*((sig/(w**2*eps**2)+1)**(0.5)-1))**(0.5)
def beta(w):
        return w*(mu*eps/2*((sig/(w**2*eps**2)+1)**(0.5)+1))**(0.5)
  
w=np.linspace(1,10**12,1000)


   
plt.plot(w,alpha(w),'b')
plt.plot(w,beta(w),'r')
plt.legend(['alpha','beta'],loc='right')
plt.xlabel("w")
plt.ylabel("alpha,beta")
plt.title("SiO2 (alpha y beta)")
plt.xlim(0,1*10**12)
plt.ylim(-100,6000)
plt.grid()
plt.show()
```

```
#Grafico del ángulo refractado y los coeficientes de Fresnel para nt/ni=1.5 y ni/nt=1.5.
import numpy as np
import matplotlib.pyplot as plt

n1=  1         #aire
n2=  1.52         #BK7
def rll(Oi):
        return (n2*np.cos(Ot) - n1*np.cos(Oi))/(n1*np.cos(Oi) + n2*np.cos(Ot))
def rT(Oi):
        return (n2*np.cos(Oi)-n1*np.cos(Ot))/(n2*np.cos(Oi) + n1*np.cos(Ot))
def tll(Oi):
        return n2*2*((n2*np.cos(Oi))/(n2*np.cos(Oi)+np.cos(Ot)))
def tT(Oi):
        return (2*n2*np.cos(Oi))/(n2*np.cos(Oi) + n1*np.cos(Ot))
def A(Oi):
  return Oi/Oi-1
Ot=0.0
Oi=np.linspace(0,2*np.pi,1000)

    
plt.plot(Oi,rll(Oi),'brown')
plt.plot(Oi,tll(Oi),'r')
plt.plot(Oi,rT(Oi),'g')
plt.plot(Oi,tT(Oi),'y')
plt.plot(Oi,A(Oi),'black')
plt.legend(['r_ll','t_ll','r_T','t_T'],loc='lower right')
plt.xlabel("Theta_i")
plt.ylabel("Coeficientes")
plt.title("Coeficientes de fresnel")
plt.xlim(0,np.pi/2)
plt.ylim(-1,1)
plt.grid()
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
from scipy import constants as cts

#Primero declaramos los símbolos a operar
n_r=sy.symbols('n_r')
n_i=sy.symbols('n_i')
A=sy.symbols('A')
B=sy.symbols('B')

#solucionamos para n_r
s1=sy.solve(n_r**4-n_r**2*A-B**2/4,n_r)
#solucionamos para n_i
s2=sy.solve(n_i**4+n_i**2*A-B**2/4,n_i)
#imprimimos las soluciones
print('n_r=',s1)
print('n_i=',s2)
```

```
import numpy as np
import matplotlib.pyplot as plt
from scipy import constants as cts

#en el visible w=10^15
w=10**15*np.arange(1,20,0.001)
w0=10**16
gamma=10**14
N=10**28
m=cts.m_e
q=cts.elementary_charge
eps=cts.epsilon_0

A= 1+(N*q**2)/(m*eps)*((w0**2-w**2)/((w0**2-w**2)**2+w**2*gamma**2))
B=(N*q**2)/(m*eps)*((w*gamma)/((w0**2-w**2)**2+w**2*gamma**2))

n_R=(A/2 + (A**2 + B**2)**0.5 /2)**0.5
n_I=(-A/2 +(A**2 + B**2)**0.5 /2)**0.5

fig, ax = plt.subplots(figsize=(7, 5), constrained_layout=True)
plt.plot(w, n_R, label='Índice de refracción $N_R$',c='r')
plt.plot(w, n_I, label='Coeficiente de extinción $N_I$',c='b')
plt.axvline(x=w0, ymin=0.0, ymax=5,label="$w_0$",c='black', linestyle='--')
plt.ylabel(' $N_R$ y $N_I$')
plt.xlabel(' $\omega$ [Hz]')
plt.title('n($\omega$)')
plt.legend()
plt.show()
```

```
#Gráfico de Cauchy y Sellmeier para materiales 
import numpy as np
import matplotlib.pyplot as plt
#longitud de onda
s = np.arange(300,900,0.01)
lamb = s*10**(-9)

A = np.zeros(3)
B = np.zeros(3)
S1 = np.zeros(9)
S2 = np.zeros(9)
Cauchy = []
sellmeier = []


#BK7
A[0] = 1.5046
B[0] = 0.00420*10**(-12)
S1[0]=1.0396
S1[1]=0.2317
S1[2]=1.01046
S2[0]=6.007*10**(-15)
S2[1]=2.00179*10**(-14)
S2[2]=103.560*10**-12

#PMMA
A[1]= 1.479
B[1]= 5.480*10**-15
S1[3]=1.179
S1[4]=0.00256
S1[5]=0.0367
S2[3]=1.02 *10**-14
S2[4]=5.42*10**-16
S2[5]=14.52*10**-16

#Fused silica
A[2] = 1.4780
B[2] = 0.000454*10**-12
S1[6]=0.6961
S1[7]=0.49794
S1[8]=0.89747
S2[6]=0.06961*10**-15
S2[7]= 0.0062*10**-12
S2[8]=0.979340*10**-8


etiquetas = ['BK7', 'PMMA','Fused silica']
#Calculamos cauchy y sellmeier
for i in (0,1,2):
  Cauchy.append((A[i] + B[i]/lamb**2))
  sellmeier.append((1+(S1[3*i]*lamb**2)/(lamb**2-S2[3*i]) +(S1[3*i+1]*lamb**2)/(lamb**2-S2[3*i+1])+(S1[3*i+2]*lamb**2)/(lamb**2-S2[3*i+2]))**0.5)

#Graficamos
fig, ax = plt.subplots(figsize=(8, 7), constrained_layout=True)
for i in (0,1,2):
  plt.plot(lamb, Cauchy[i], label=etiquetas[i])
  plt.plot(lamb, sellmeier[i], label=etiquetas[i], linestyle='--')

plt.ylabel('Indice de refracción')
plt.xlabel('Longitud de onda $\lambda$ [$m$]')
plt.title('Gráfica de dispersión n($\lambda$)')
#plt.xlim(0,100)
plt.legend()
plt.show()
```

```
#Gráfico de n para metales, Plata
import numpy as np
import matplotlib.pyplot as plt
from scipy import constants as cts

#en el visible w=10^15
w=10**15*np.arange(1,30,0.001)

gamma=4100
wp=1.63*10**16
m=cts.m_e


F= 1-(wp**2)/(w**2+gamma**2)
G=(gamma*wp**2)/(w*(w**2+gamma**2))

n_R=(F/2 + (F**2 + G**2)**0.5 /2)**0.5
n_I=(-F/2 +(F**2 + G**2)**0.5 /2)**0.5

fig, ax = plt.subplots(figsize=(7, 5), constrained_layout=True)
plt.plot(w, n_R, label='Índice de refracción $n_R$',c='r')
plt.plot(w, n_I, label='Coeficiente de extinción $n_I$',c='b')
plt.axvline(x=wp, ymin=0.0, ymax=5,label="$w_p$",c='black', linestyle='--')
plt.ylabel(' $n_R$ y $n_I$')
plt.xlabel(' $\omega$ [Hz]')
plt.title('n($\omega$) para Metales, Plata')
plt.ylim(-1,3)
plt.legend()
plt.show()
```
